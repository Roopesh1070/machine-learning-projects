```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowlotfrontage,upplotfrontage=remove_outlier(df['mpg'])
df["mpg"]=np.where(df['mpg']>upplotfrontage,upplotfrontage,df['mpg'])
df["mpg"]=np.where(df['mpg']<lowlotfrontage,lowlotfrontage,df['mpg'])
```

```python
lowlotfrontage,upplotfrontage=remove_outlier(df['horsepower'])
df["horsepower"]=np.where(df['horsepower']>upplotfrontage,upplotfrontage,df['horsepower'])
df["horsepower"]=np.where(df['horsepower']<lowlotfrontage,lowlotfrontage,df['horsepower'])
```
